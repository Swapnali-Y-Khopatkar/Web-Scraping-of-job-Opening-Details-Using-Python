```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
URL = "https://www.freshersworld.com/premium-jobs?src=homeheader"
```

```python
import requests
r = requests.get(URL)
htmlcontent = r.content
htmlcontent
```

```python
soup = BeautifulSoup(htmlcontent,"html.parser")
```

```python
Job_titles= soup.find_all("span",class_="wrap-title seo_title")
print(Job_titles)
```

```python
list_Job_Titles=[]
for i in range(0,len(Job_titles)):
    list_Job_Titles.append(Job_titles[i].get_text())    
print(list_Job_Titles)
```

```python
company_names= soup.find_all(class_="latest-jobs-title font-16 margin-none inline-block company-name")
print(company_names)
```

```python
list_company_names=[]
for i in range(0,len(company_names)):
    list_company_names.append(company_names[i].get_text())
print(list_company_names)
```

```python
Location= soup.find_all(class_="bold_font")
print(Location)
```

```python
list_Location=[]
for i in range(0,len(company_names)):
    list_Location.append(Location[i].get_text())

print(list_Location)
```

```python
Qualifications= soup.find_all(class_="qualifications display-block modal-open pull-left job-details-span")
print(Qualifications)
```

```python
list_Qualifications=[]
for i in range(0,len(Qualifications)):
    list_Qualifications.append(Qualifications[i].get_text())
print(list_Qualifications)
```

```python
import pandas as pd
df=pd.DataFrame({"Job Titles":list_Job_Titles,"Company Names":list_company_names,"Location":list_Location,
                 "Qualifications":list_Qualifications})
df
```

```python
df.to_csv("JobReading data.csv",index=False)
a=pd.read_csv("JobReading data.csv")
a
```
